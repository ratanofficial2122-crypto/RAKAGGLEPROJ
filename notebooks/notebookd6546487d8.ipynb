```python
# -------------------------------
# Step 0: Install required packages
# -------------------------------
!pip install python-docx PyGithub --quiet

# -------------------------------
# Step 1: Import libraries
# -------------------------------
import pandas as pd
from docx import Document
import os

# -------------------------------
# Step 2: Create working directory
# -------------------------------
os.makedirs("/kaggle/working/", exist_ok=True)

# -------------------------------
# Step 3: Create sample CSV dataset
# -------------------------------
data = [
    {"Title": "AI YouTube Agent", "Description": "Automates channel growth", "Views": 1000},
    {"Title": "Python Tutorial", "Description": "Learn Python basics", "Views": 500},
    {"Title": "Fun Shorts", "Description": "Funny short video", "Views": 2000},
]

df = pd.DataFrame(data)
csv_path = "/kaggle/working/sample_youtube_dataset.csv"
df.to_csv(csv_path, index=False)

# -------------------------------
# Step 4: Create DOCX summary
# -------------------------------
doc = Document()
doc.add_heading("Project Summary", level=1)
doc.add_paragraph("This is a sample YouTube dataset project. Details:")
for i, row in df.iterrows():
    doc.add_paragraph(f"{i+1}. Title: {row['Title']}, Description: {row['Description']}, Views: {row['Views']}")

docx_path = "/kaggle/working/project_summary.docx"
doc.save(docx_path)

# -------------------------------
# Step 5: Create README.md
# -------------------------------
readme_path = "/kaggle/working/README.md"
with open(readme_path, "w") as f:
    f.write("# Kaggle Capstone Project\n")
    f.write("This repository contains a sample YouTube dataset project.\n")
    f.write("- sample_youtube_dataset.csv: CSV dataset\n")
    f.write("- project_summary.docx: Project summary\n")

# -------------------------------
# Step 6: Verify files
# -------------------------------
print("Files created in /kaggle/working/:")
print(os.listdir("/kaggle/working/"))

# -------------------------------
# Optional Step 7: Upload to GitHub
# -------------------------------
# Uncomment and configure below if you want to upload to GitHub directly
"""
from github import Github

token = "PASTE_YOUR_GITHUB_PAT_HERE"
repo_name = "ratanofficial2122-crypto/RAKAGGLEPROJ"
branch = "main"

g = Github(token)
repo = g.get_repo(repo_name)

for file_name in os.listdir("/kaggle/working/"):
    file_path = os.path.join("/kaggle/working/", file_name)
    if os.path.isfile(file_path):
        with open(file_path, "rb") as f:
            content = f.read()
        try:
            contents = repo.get_contents(file_name, ref=branch)
            repo.update_file(contents.path, f"Update {file_name}", content, contents.sha, branch=branch)
            print(f"Updated {file_name}")
        except:
            repo.create_file(file_name, f"Add {file_name}", content, branch=branch)
            print(f"Created {file_name}")
"""
```
