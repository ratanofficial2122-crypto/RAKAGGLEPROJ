```python
#--------------------------
#Import + Define Agents
#--------------------------

import concurrent.futures
import time
import os

# Create working folder
os.makedirs("/kaggle/working/", exist_ok=True)

# ---------------------------
# Define "agents"
# ---------------------------

def agent_ideas():
    time.sleep(1)
    return [
        "How I Automated My YouTube Channel Using AI",
        "Top 5 AI Tools Every Creator Should Use",
        "Make Viral Shorts Using ChatGPT"
    ]

def agent_description():
    time.sleep(1)
    return [
        "This video explains how AI tools automate YouTube workflows.",
        "Learn top AI tools to grow your channel fast!",
        "Create viral shorts using powerful AI prompts."
    ]

def agent_hashtags():
    time.sleep(1)
    return [
        "#AI #YouTubeAutomation #ChatGPT #Creators",
        "#AITools #YouTubeTips #Automation",
        "#Shorts #ViralVideo #AIShorts"
    ]
print ("imported concurrent.future time os and defined agent sucessfully")
```

imort concurrent.future time and os and defined agent

```python
# ---------------------------
# Run agents in parallel
# ---------------------------

with concurrent.futures.ThreadPoolExecutor() as executor:
    future_ideas = executor.submit(agent_ideas)
    future_desc = executor.submit(agent_description)
    future_tags = executor.submit(agent_hashtags)

    ideas = future_ideas.result()
    descriptions = future_desc.result()
    hashtags = future_tags.result()
print ("run agent and assign result to variables sucessfully")
```

run agents in parralel

```python
# ---------------------------
# Combine & save and the outputs
# ---------------------------

combined_output = ""
for i in range(3):
    combined_output += f"Video {i+1}:\n"
    combined_output += f"  Title: {ideas[i]}\n"
    combined_output += f"  Description: {descriptions[i]}\n"
    combined_output += f"  Hashtags: {hashtags[i]}\n"
    combined_output += "\n"

# Save to file
output_path = "/kaggle/working/youtube_multi_agent_output_parallel.txt"

with open(output_path, "w") as f:
    f.write(combined_output)

print("Parallel Agent Output File Created!")
print("File Path:", output_path)
print("\nGenerated Output:\n")
print(combined_output)
```
